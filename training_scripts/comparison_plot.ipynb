```python
import matplotlib.pyplot as plt
import seaborn as sns

accuracies = {
    'Naive Bayes': 0.9668064709406831,      
    'Logistic Regression': 0.9841821449970042, 
    'KNN (Sampled)': 0.7525 
}

models = list(accuracies.keys())
values = list(accuracies.values())

plt.figure(figsize=(10, 6))
colors = ['#3498db', '#2ecc71', '#e67e22']

sns.barplot(x=models, y=values, palette=colors)

plt.xlabel('Machine Learning Algorithms', fontsize=12)
plt.ylabel('Accuracy Score', fontsize=12)
plt.title('Final Model Performance Comparison', fontsize=14)
plt.ylim(0.7, 1.0) 

for i, v in enumerate(values):
    plt.text(i, v + 0.005, f"{v*100:.2f}%", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
